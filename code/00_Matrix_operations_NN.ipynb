### Single batch (batch_size=1)

Single batch (1 observation) with three inputs (features).

```python
x1 = 10
x2 = 5
x3 = 6

w1 = 0.1
w2 = 0.2
w3 = 0.3

b = 1
```

The basic linear operation : $\sum_ix_iw_i + b$

```python
sigma = x1*w1 + x2*w2 + x3*w3 + b
print(sigma)
```

Using the matrices

```python
X = np.array([x1,x2,x3])
W = np.array([w1,w2,w3])
```

**When the two vectors are 1D, np.dot() makes the scalar product !**

```python
print(np.dot(X, W) + b)
print(np.dot(X, W.T) + b)
```

The non linear operation

```python
# The non-linear operation
y_hat = np.where(sigma > 0, sigma, 0)
print(y_hat)
```

### Multiple batch (batch_size=3)

Same Neural Network but this time we input multiple batch (e.g. 2 observations) with three inputs (features) each.

```python
# First batch
x11 = 1
x12 = 5
x13 = 6
# Second batch
x21 = 10
x22 = 53
x23 = 62
# Single unit
w1 = 0.1
w2 = 0.2
w3 = 0.3

b = 1
```

The basic linear operation : $\sum_ix_iw_i + b$

```python
sigma = x11*w1 + x12*w2 + x13*w3 + b, x21*w1 + x22*w2 + x23*w3 + b
print(sigma)
```

Using the matrices

```python
X = np.array([[x11,x12,x13],
              [x21,x22,x23]])
print(X)

W = np.array([w1,w2,w3])
print(W)
```

**With multiple batch, we transpose W in the .dot product**

```python
hidden_output = np.dot(X, W.T) + b

print(hidden_output)
```

The non linear operation

```python
# The non-linear operation
y_hat = np.where(hidden_output > 0, sigma, 0)

print(y_hat)
```

## Combining units gives a layer

Vertical combination of units gives a layer. Below an hidden layer composed of 3 units.  

```python
# The two inputs features:
x1 = 10
x2 = 5

target = 0.9

# Input weights and bias of the hidden layer:
# The two weights of the first Neuron of the input hidden layer
win_11 = 0.4
win_12 = 0.2
# The two weights of the second Unit of the input hidden layer
win_21 = 0.3
win_22 = 0.3
# The two weights of the third Node of the input hidden layer
win_31 = 0.1
win_32 = 0.5

# The same bias for every Unit
b_in = 1

# Output weights and bias of the hidden layer:
# The two weights of the first Neuron of the output hidden layer
wout_1 = 0.4
wout_2 = 0.2
wout_3 = 0.2

# The same bias for every Unit
b_out = 1
```

```python
print(x1*win_11 + x2*win_12 + b)
print(x1*win_21 + x2*win_22 + b)
print(x1*win_31 + x2*win_32 + b)
```

Let's put it into a numpy operation

```python
X = np.array([x1,x2])

print(X)
print(X.shape)
```

```python
# The weights of the input hidden layer:
Win = np.array([[win_11,win_12],[win_21,win_22],[win_31,win_32]])

print(Win)
```

```python
# Linear transformation of the input hidden layer
in_hidden = np.dot(X, Win.T) + b
print(in_hidden)
```

```python
# Non Linear transformation (ReLu) of the inputs:
phi = np.where(in_hidden > 0, in_hidden, 0)
print(phi)
```

```python
y_hat = phi[0]*wout_1 + phi[1]*wout_2 + phi[2]*wout_3 + b_out

print(y_hat)
```

```python
Wout = np.array([wout_1, wout_2, wout_3])
```

```python
y_hat = np.dot(phi, Wout.T)

print(y_hat)
```
